**CNN de resimlerin kare formatinda olmasinin nedeni nedir?**

CNN (Convolutional Neural Network) kullanrken ilk resimlerin kare formatnda olmasnn temel nedeni, arlkl olarak kullanlan konvolsyon ilemi ve havuzlama ilemi iin bir kare matrisin daha uygun olmasdr. te bu tercihin temel nedenleri:

**lemlerin Simetrisi:** Konvolsyon ilemi, resmin farkl blgelerini ayn filtre ile ileyerek zellikleri karmak iin kullanlr. Kare bir matris, ilemlerin simetrik olmasn salar ve bylece daha tutarl sonular elde edilir.

**Daha Kolay Matematik:** Kare matrisler, matematiksel ilemler asndan daha kolaydr. Filtre boyutu, adm bykl ve havuzlama ilemi boyutu gibi parametreleri ayarlamak daha basittir.

**Genelletirme Yetenei:** Kare resimler, daha genel ve eitli zelliklerin renilmesini kolaylatrabilir. Kare bir resim, daha farkl ynlerden gelen zellikleri ierebilir ve bu da modelin daha iyi genelleme yapmasna yardmc olabilir.

**Veri Uygunluu:** Birok veri kayna, zellikle eitim verileri, genellikle kare formatnda resimler ierir. Bu nedenle, veri uyumluluunu korumak iin kare resimler kullanmak daha uygun olabilir.

Ancak, bu **kare resim** kural **mutlak bir gereklilik deildir**. Dikdrtgen veya farkl boyutlardaki resimler de kullanlabilir, ancak bu durumda an yaplandrlmas ve parametrelerin ayarlanmas daha karmak olabilir. Genellikle, veri ve probleme zg olarak resim boyutlar seilir, ancak **kare resimler sklkla tercih edilir** (best practice).

**CNN de test datasina da img data generator uygulanir mi?**  ----------- **UYGULANMASA DAHA IYI OLUR (UDEMY'den)**

Evet, CNN (Convolutional Neural Network) veya baka bir derin renme modeli kullanrken test verilerine de veri artrma (data augmentation) uygulanabilir. Veri artrma, eitim verilerini eitlendirmek ve modelin daha iyi genelleme yapmasn salamak iin kullanlan bir tekniktir. Ancak test verileri zerinde veri artrma uygulanmas, genellikle eitim verileri zerinde uyguland gibi youn bir ekilde yaplmaz. Bunun yerine, test verilerini daha iyi deerlendirebilmek iin **hafif bir ekilde kullanlabilir**.

Test verileri iin veri artrmann baz avantajlar ve dikkate alnmas gerekenler unlar olabilir:

**Daha yi Deerlendirme:** Test verileri zerinde hafif bir veri artrma uygulamak, modelin daha iyi bir ekilde genelleme yapmasna yardmc olabilir. zellikle modelin eitim srasnda grd zelliklerin varyasyonlarn test verilerinde deerlendirmek istiyorsanz bu kullanl olabilir.

**Tutarszlk Kontrol:** Veri artrma, bazen test verilerindeki tutarszlklar dengelemeye yardmc olabilir. rnein, eitim verilerinde ok fazla bir snfn rnekleri bulunurken dier snflarn rnekleri azsa, veri artrma ile bu dengesizlii hafifletebilirsiniz.

**Hafif Veri Artrma:** Test verileri zerinde uygulanacak veri artrma, eitim verilerine kyasla daha hafif olmaldr. Bu, test verilerini bozmamak ve gerek dnya senaryolarn daha iyi yanstmak iin nemlidir.

**rnein,** bir test resmi iin rastgele yatay veya dikey dndrme, hafif bir kesme veya renk deiiklikleri gibi hafif veri artrma teknikleri kullanabilirsiniz. Ancak bu, test verilerinin aslna sadk kalmasn salayacak ekilde dikkatli bir ekilde yaplmaldr.

**Sonu olarak,** test verilerine de veri artrma uygulayabilirsiniz, ancak bu ilem eitim verileri zerindeki kadar youn olmamal ve test verilerinin doruluunu bozmamaldr. Ayrca, modelin gerek dnya senaryolarn daha iyi deerlendirmesine yardmc olacak ekilde seici ve dikkatli bir ekilde yaplmaldr.

# Creating the Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dropout, Flatten, Dense, Conv2D, MaxPooling2D
# Avgpooling2D de  deneyebilir, sonuclari karsilastirabilirsin.
```

```python
# Bunun adi mimari yapi (architecture)

model = Sequential()

model.add(Conv2D(filters=64, kernel_size=(3,3),input_shape=image_shape, padding='same', activation='relu',))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=128, kernel_size=(3,3), padding='same', activation='relu',))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=128, kernel_size=(3,3), padding='same', activation='relu',))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=256, kernel_size=(3,3), padding='same', activation='relu',)) # filtre sayisini azalt ki parametre sayisi azalsin. 
model.add(Conv2D(filters=256, kernel_size=(3,3), padding='same', activation='relu',)) # Filtre sayisi olaak 256 verilmesi tavsiye edimez. 
model.add(Conv2D(filters=256, kernel_size=(3,3), padding='same', activation='relu',)) # 256 olunca parametre cok artar.
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=128, kernel_size=(3,3), padding='same', activation='relu',))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=64, kernel_size=(3,3), padding='same', activation='relu',))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())

#---ANN----

model.add(Dense(128)) # parametreleri artirmamak adina tek hidden layer kullandik ama yetmedi. Cok parametre var.
model.add(Activation('relu'))

model.add(Dropout(0.5))

model.add(Dense(1))
model.add(Activation('sigmoid')) # 2 class oldugu icin Sigmoid kullandik !!!

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```
