**CNN de resimlerin kare formatinda olmasinin nedeni nedir?**

CNN (Convolutional Neural Network) kullanırken ilk resimlerin kare formatında olmasının temel nedeni, ağırlıklı olarak kullanılan konvolüsyon işlemi ve havuzlama işlemi için bir kare matrisin daha uygun olmasıdır. İşte bu tercihin temel nedenleri:

**İşlemlerin Simetrisi:** Konvolüsyon işlemi, resmin farklı bölgelerini aynı filtre ile işleyerek özellikleri çıkarmak için kullanılır. Kare bir matris, işlemlerin simetrik olmasını sağlar ve böylece daha tutarlı sonuçlar elde edilir.

**Daha Kolay Matematik:** Kare matrisler, matematiksel işlemler açısından daha kolaydır. Filtre boyutu, adım büyüklüğü ve havuzlama işlemi boyutu gibi parametreleri ayarlamak daha basittir.

**Genelleştirme Yeteneği:** Kare resimler, daha genel ve çeşitli özelliklerin öğrenilmesini kolaylaştırabilir. Kare bir resim, daha farklı yönlerden gelen özellikleri içerebilir ve bu da modelin daha iyi genelleme yapmasına yardımcı olabilir.

**Veri Uygunluğu:** Birçok veri kaynağı, özellikle eğitim verileri, genellikle kare formatında resimler içerir. Bu nedenle, veri uyumluluğunu korumak için kare resimler kullanmak daha uygun olabilir.

Ancak, bu **kare resim** kuralı **mutlak bir gereklilik değildir**. Dikdörtgen veya farklı boyutlardaki resimler de kullanılabilir, ancak bu durumda ağın yapılandırılması ve parametrelerin ayarlanması daha karmaşık olabilir. Genellikle, veri ve probleme özgü olarak resim boyutları seçilir, ancak **kare resimler sıklıkla tercih edilir** (best practice).

CNN (Convolutional Neural Network) modelleri oluştururken Image Data Generator kullanmak genellikle iyi bir uygulamadır. Image Data Generator, özellikle büyük veri kümeleri üzerinde çalışırken ve veri artırma (data augmentation) işlemleri yapmak istediğinizde çok faydalıdır. İşte Image Data Generator'ın bazı avantajları:

Veri Artırma: Veri Artırma, mevcut eğitim veri kümenizi çeşitlendirmek ve modelinizi daha genelleştirilebilir hale getirmek için kullanılan bir tekniktir. Image Data Generator, resimleri rastgele döndürme, yeniden boyutlandırma, yatay veya dikey olarak aynalama, parlaklık değişimi gibi işlemlerle artırabilir. Bu, modelinizin daha fazla çeşitlilikle eğitilmesini sağlar.

Veri Yükünü Azaltma: Image Data Generator, bellek verimliliğini artırmak için resimleri yükleme işlemi sırasında gerçek zamanlı olarak dönüştürebilir. Bu, büyük veri kümeleriyle çalışırken işlemci ve bellek kaynaklarını daha etkili bir şekilde kullanmanıza yardımcı olur.

Mini-Yığınlar (Mini-Batches) Oluşturma: Image Data Generator, eğitim sırasında veriyi mini-yığınlar halinde (mini-batches) otomatik olarak oluşturabilir. Bu, büyük veri kümelerinin daha iyi işlenmesini sağlar ve aynı zamanda eğitim sürecini hızlandırabilir.

Kolay Kullanım: Image Data Generator, Keras ve TensorFlow gibi yaygın kullanılan derin öğrenme çerçeveleriyle uyumlu bir şekilde kullanılabilir. Bu, modelinizi oluştururken ve eğitirken kolayca entegre edilebilir.

Ancak Image Data Generator'ı kullanırken dikkate almanız gereken bazı önemli hususlar vardır. Örneğin, veri artırma işlemleri, veriye özgü olarak yapılmalıdır ve bazı durumlarda fazla veri artırma modelinizi kötüleştirebilir. Ayrıca, bazı uygulamalarda Image Data Generator'ı kullanmadan önce resimleri elle ön işlemeniz gerekebilir.

Sonuç olarak, Image Data Generator, CNN modelleri oluştururken veri artırma ve veri yükünü azaltma işlemleri için güçlü bir araçtır ve genellikle iyi bir uygulamadır. Ancak veri kümenizin özelliklerine ve projenizin gereksinimlerine bağlı olarak kullanımını dikkatlice planlamalısınız.

ImageDataGenerator sınıfı, Keras kütüphanesinde yer alan ve görüntü verilerini artırarak daha fazla veri üretmek için kullanılan bir veri artırma aracıdır.

Bu kod bloğu, ImageDataGenerator sınıfından bir örnek oluşturur ve bu örneğe veri artırma yöntemlerinin parametrelerini belirtir. Aşağıdaki parametreler kullanılır:

rotation_range: Rastgele döndürme açısı aralığı (0-180 derece arasında).

width_shift_range: Rastgele yatay kaydırma aralığı, görüntü genişliğinin yüzdesi olarak belirtilir.

height_shift_range: Rastgele dikey kaydırma aralığı, görüntü yüksekliğinin yüzdesi olarak belirtilir.

shear_range: Rastgele kayma aralığı, kayma açısı derecesi olarak belirtilir.

zoom_range: Rastgele yakınlaştırma aralığı.

horizontal_flip: Rastgele yatay çevirme yapar.

fill_mode: Görüntü genişletilirken veya küçültülürken oluşan boş piksellerin doldurulma yöntemi. 'nearest' yöntemi, piksellerin en yakın komşusunun rengi ile doldurulmasını sağlar.

Bu parametreler, image_gen adlı bir veri artırma nesnesi oluşturur ve daha sonra modelin eğitimi sırasında bu nesne kullanılarak veri artırma yapılabilir. Bu, modelin daha fazla veriye erişmesine ve genelleştirme yapmasına yardımcı olur.

```python
# Computer visionda bu cok kullaniliyor. Orada resim sayisi artacak. CNN'de sayi artmiyor,sekil degisiyor, bu sayede resim sayisi,  Epoch sayisina bagli olarak, 100 kat bile olabiliyor. 
image_gen = ImageDataGenerator(rotation_range=15,  
                               width_shift_range=0.10, 
                               height_shift_range=0.10, 
                               #rescale=1/255,   
                               shear_range=0.1, 
                               zoom_range=0.1, 
                               horizontal_flip=True, 
                               fill_mode='nearest')
# bunlar max degerler buraya kadar degistir her seferinde diyoruz. Bu degeri cok de buyutmek dogru degil. 
# resimler öncesinden rescale edilmiş olduğundan tekrar rescale yapmadık.
# Her sefer baska birini kullaniyor bu adimlardan. 

```


ImageDataGenerator ile veri artırma işlemi, var olan veri kümesindeki görüntüleri manipüle ederek yeni görüntüler oluşturur. Bu nedenle, eğitim veri setindeki veri sayısı artmaz, ancak modelin eğitim sırasında kullanabileceği daha fazla örnek varmış gibi davranır.

Örneğin, eğer eğitim veri setinde 1000 adet görüntü varsa ve ImageDataGenerator'ı belirtilen parametrelerle kullanarak her görüntüden 10 yeni görüntü oluşturulursa, toplam eğitim verisi sayısı 10000 olur. Ancak bu yeni verileri diskte depolamak yerine, modelin eğitimi sırasında dinamik olarak üretilir ve bellekte tutulur. **CNN de test datasina da img data generator uygulanir mi?**  ----------- **UYGULANMASA DAHA IYI OLUR (UDEMY'den)**

Evet, CNN (Convolutional Neural Network) veya başka bir derin öğrenme modeli kullanırken test verilerine de veri artırma (data augmentation) uygulanabilir. Veri artırma, eğitim verilerini çeşitlendirmek ve modelin daha iyi genelleme yapmasını sağlamak için kullanılan bir tekniktir. Ancak test verileri üzerinde veri artırma uygulanması, genellikle eğitim verileri üzerinde uygulandığı gibi yoğun bir şekilde yapılmaz. Bunun yerine, test verilerini daha iyi değerlendirebilmek için **hafif bir şekilde kullanılabilir**.

Test verileri için veri artırmanın bazı avantajları ve dikkate alınması gerekenler şunlar olabilir:

**Daha İyi Değerlendirme:** Test verileri üzerinde hafif bir veri artırma uygulamak, modelin daha iyi bir şekilde genelleme yapmasına yardımcı olabilir. Özellikle modelin eğitim sırasında gördüğü özelliklerin varyasyonlarını test verilerinde değerlendirmek istiyorsanız bu kullanışlı olabilir.

**Tutarsızlık Kontrolü:** Veri artırma, bazen test verilerindeki tutarsızlıkları dengelemeye yardımcı olabilir. Örneğin, eğitim verilerinde çok fazla bir sınıfın örnekleri bulunurken diğer sınıfların örnekleri azsa, veri artırma ile bu dengesizliği hafifletebilirsiniz.

**Hafif Veri Artırma:** Test verileri üzerinde uygulanacak veri artırma, eğitim verilerine kıyasla daha hafif olmalıdır. Bu, test verilerini bozmamak ve gerçek dünya senaryolarını daha iyi yansıtmak için önemlidir.

**Örneğin,** bir test resmi için rastgele yatay veya dikey döndürme, hafif bir kesme veya renk değişiklikleri gibi hafif veri artırma teknikleri kullanabilirsiniz. Ancak bu, test verilerinin aslına sadık kalmasını sağlayacak şekilde dikkatli bir şekilde yapılmalıdır.

**Sonuç olarak,** test verilerine de veri artırma uygulayabilirsiniz, ancak bu işlem eğitim verileri üzerindeki kadar yoğun olmamalı ve test verilerinin doğruluğunu bozmamalıdır. Ayrıca, modelin gerçek dünya senaryolarını daha iyi değerlendirmesine yardımcı olacak şekilde seçici ve dikkatli bir şekilde yapılmalıdır.

# Creating the Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dropout, Flatten, Dense, Conv2D, MaxPooling2D
# Avgpooling2D de  deneyebilir, sonuclari karsilastirabilirsin.
```

```python
# Bunun adi mimari yapi (architecture)

model = Sequential()

model.add(Conv2D(filters=64, kernel_size=(3,3),input_shape=image_shape, padding='same', activation='relu',))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=128, kernel_size=(3,3), padding='same', activation='relu',))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=128, kernel_size=(3,3), padding='same', activation='relu',))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=256, kernel_size=(3,3), padding='same', activation='relu',)) # filtre sayisini azalt ki parametre sayisi azalsin. 
model.add(Conv2D(filters=256, kernel_size=(3,3), padding='same', activation='relu',)) # Filtre sayisi olaak 256 verilmesi tavsiye edimez. 
model.add(Conv2D(filters=256, kernel_size=(3,3), padding='same', activation='relu',)) # 256 olunca parametre cok artar.
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=128, kernel_size=(3,3), padding='same', activation='relu',))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=64, kernel_size=(3,3), padding='same', activation='relu',))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())

#---ANN----

model.add(Dense(128)) # parametreleri artirmamak adina tek hidden layer kullandik ama yetmedi. Cok parametre var.
model.add(Activation('relu'))

model.add(Dropout(0.5))

model.add(Dense(1))
model.add(Activation('sigmoid')) # 2 class oldugu icin Sigmoid kullandik !!!

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

image_gen.flow_from_directory() yöntemi, verilerin diskteki bir dizindeki alt dizinlerde sınıflandırılmış olduğu bir klasör yapısını kullanarak, eğitim verileri için bir veri akışı oluşturur.

Bu yöntemin parametreleri şöyledir:

directory: Eğitim verilerinin bulunduğu dizin yolu.
target_size: Görüntülerin yeniden boyutlandırılacağı hedef boyut.
color_mode: Görüntü renk kanallarının modu, örn. "rgb" veya "grayscale".
batch_size: Tek seferde oluşturulacak veri yığınının boyutu.
class_mode: Sınıflandırma yöntemi, örn. "binary" veya "categorical".
shuffle: Verilerin karıştırılıp karıştırılmayacağı.
seed: Verilerin karıştırılmasında kullanılacak rastgele sayı üretecinin önceden belirlenmiş bir başlangıç değeri.
Bu yöntem, her çağrıldığında bir "batch" veri üretir, böylece eğitim sırasında modelin her adımda yeni verilerle çalışması sağlanır. Ayrıca, ImageDataGenerator ile belirlenen veri artırma işlemleri, her çağrıda dinamik olarak uygulanır, böylece eğitim sırasında model daha fazla veriye erişir ve daha iyi genelleştirme yapabilir.
